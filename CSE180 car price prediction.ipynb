```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("CarPrice_Assignment.csv")
df
```

```python
df["fueltype"].replace({"gas":1, "diesel":0}, inplace=True)
```

```python
df
```

```python
df["drivewheel"].replace({"rwd":1, "4wd":0, "fwd":2}, inplace=True)
```

```python
df
```

```python
df.isnull().value_counts()
```

```python
df.columns
```

```python
df["enginelocation"].unique()
```

```python
df["enginelocation"].replace({"front":1, "rear":0}, inplace=True)
```

```python
df
```

```python
#independent variable 
x=df[['boreratio', 'stroke','compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg','fueltype','drivewheel','enginelocation',]]

#dependent variable
y=df["price"] 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
x_test.index
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin=LinearRegression()
```

```python
lin.fit(x_train,y_train)
```

```python
p=lin.predict(x_test)
p
```

```python
d=pd.DataFrame({"Predicted":p, "Actual":y_test})
d
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,p)
```
